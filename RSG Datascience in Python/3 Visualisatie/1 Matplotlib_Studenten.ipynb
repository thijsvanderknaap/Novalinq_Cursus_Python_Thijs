# Procedural programmeren/ Procedural programming  
  
  Functional programming is de naam voor een programmeer-methode.  
  Een paradigma dus.

  Met elke functie als apart <em>schakeltje</em> in de richting van de output.

# Functioneel programmeren/ Functional programming  
  
  Functional programming is de naam voor een programmeer-methode.  
  Een paradigma dus.  
  
  Alles draait om de <strong>puurheid</strong> van de executie.  
  <strong>Functies</strong> en <strong>data</strong> worden volledig gescheiden.  
    
  De <strong>uitkomst</strong> van een functie is zo dus afhankelijk van de <strong>input.</strong>  

# Object georienteerd programmeren/ Object Oriented programming  
  
  Object Oriented programming of OOP (for short) is de naam voor een programmeer-methode.  
  Een paradigma dus.  
    
  OOP is handig handig voor reproduceerbare code en zie je vaker bij development-werk  
  
  Je behandeld de data-behandelingen met objecten die je schrijft en aanroept.   
  In de objecten zitten <strong>attributes</strong> en <strong>methods</strong>  
  
  Attributes:  
  Parameters, aan te passen in the fly.  
    
  Methods:  
  Functie-toebehoren van het object.  
    
  Natuurlijk blijven dit schakeltjes in de grote pipeline,  
  Maar de parameters kunnen onderweg herbruikt en aangepast worden.  
  Hierdoor is de output van een object.method afhenkelijk van beide:  
  - De input  
    
  - De parameters

# Subplot.  
  
  met subplot() kunnen we meerdere plots maken.
  Binnen een gedefinieerde tabel.  
  
  Subplot() neemt:
  - nrows: Nummer van rijen  
  - ncols: Nummer van kollommen  
  - index: Plaats in de tabel
  

```python
plt.subplot(1,2,1)
plt.plot(x, y ,color= 'r')
plt.xlabel("Minutes inside")
plt.ylabel("Level of anxiety")
plt.title("Sharktank")
plt.subplot(1,2,2)
plt.plot(x, z, color='g')
plt.xlabel("Minutes inside")
plt.ylabel("Executive Precision")
plt.title("Workbench 5")
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r')
plt.xlabel("Minutes inside")
plt.ylabel("Level of anxiety")
plt.title("Sharktank")
plt.subplot(2,2,4)
plt.plot(x, z, 'b*-', markevery=4)
plt.xlabel("Minutes inside")
plt.ylabel("Executive Precision")
plt.title("Workbench 5")
plt.tight_layout()
```

# Object Oriented visualiseren:  
  
  Maak een object door het aan te roepen en te declareren.  
  Dan methodes en attributes gebruiken.
  
  - Plaatsing van meerdere visualisaties (op Axes) in 1 canvas (figure).  
  - Algeheel betere controle door plots & canvas als afzonderlijke objecten te behandelen.

```python
fig = plt.figure(figsize=(5,5), dpi=100)

axes = fig.add_axes([0, 0, 1, 1]) # links, onder, wijdte, hoogte (range 0 tot 1)

axes.plot(x, y, 'b')
axes.set_xlabel('Hoeveelheden aangekocht', **cast) # Notice the use of set_ to begin methods
axes.set_ylabel('x adds bekeken', **cast)
axes.set_title('Manipulation succes', **hfont)

axes1 = fig.add_axes([0.12, 0.55, 0.4, 0.4]) # links, onder, wijdte, hoogte (range 0 tot 1)

axes1.plot(x, z, 'r')
axes1.set_xlabel('Reclame bekeken') # Notice the use of set_ to begin methods
axes1.set_ylabel('enthusiasme merk')
```
