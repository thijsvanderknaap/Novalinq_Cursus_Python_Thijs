# Meer Corr-stylings!  
  
  En meer dan het styleren van je correlatieplot.  
  
  Pandas style functie geeft onwijs veel toffe mogelijkheden  
  Voor het inkleuren van je matrixen, dus ook je dataframes.
  
  Voor meer styling mogelijkheden, zie:  
https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html



```python
def color_value(value):

    if (value < 0 and value > -1):
        color = 'red'
    elif (value > 0 and value < 1):
        color = 'green'
    else:
        color = 'black'

    return 'color: %s' % color
```

```python
corr.style.applymap(color_value)
```

```python
corr.style.background_gradient(cmap=cmap)
```
