###### Figuur 1
Werkt goed omdat er niet te veel verschillende datapunten zijn waardoor de heatmap leesbaar blijft.

We kunnen zien dat de survivalrate voor vrouwen lager is wanneer de familiegrote groter wordt (4+)
Vooral voor vrouwen is dit goed te zien.

Bij mannen is het beeld minder éénduidig, de survivalrate gaat eerst omhoog wanneer de family grote wordt, van 1 t/m 4, daarna wordt deze minder (0) bij 5 en 6, bij 7 is deze weer iets groter om daarna weer naar 0 te zakken.

Daarnaast zien we dat survival rate van mannen lager is dan dat van vrouwen.

# Meer Corr-stylings!  
  
  En meer dan het styleren van je correlatieplot.  
  
  Pandas style functie geeft onwijs veel toffe mogelijkheden  
  Voor het inkleuren van je matrixen, dus ook je dataframes.
  
  Voor meer styling mogelijkheden, zie:  
https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html



```python
def color_value(value):

    if (value < 0 and value > -1):
        color = 'red'
    elif (value > 0 and value < 1):
        color = 'green'
    else:
        color = 'black'

    return 'color: %s' % color
```

```python
corr.style.applymap(color_value)
```

```python
corr.style.background_gradient(cmap=cmap)
```

# Actie!

Schrijf een functie die zichtbaar maakt wanneer de kans op overleven groter is dan 60 procent in groen en kleiner dan 20% in rood, anders zwart. Groepeer op Pclass en Sex. 

Als je alleen wilt conditioneren op bijvoorbeeld `Survived` dan moet je de parameter `subset=['Survived']` toevoegen.

```python
# defineren van styling
def surv_value(value):

    if (value > 0.6):
        color = 'red'
    elif (value < 0.2):
        color = 'green'
    else:
        color = 'black'

    return 'color: %s' % color

# Groeperen op Pclass en sex voor het gemiddelde van survived
tit_grp_pclass_sex = TVis.groupby(['Pclass','Sex'])['Survived'].mean()

# Dit in een frame plaatsen
tit_grp_pclass_sex = tit_grp_pclass_sex.to_frame()

# toevoegen van de styling surv_value die hierboven is gedifineerd
tit_grp_pclass_sex.style.applymap(surv_value)
```

```python
survived = 'survived'
not_survived = 'not survived'
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10, 4))
women = TVis[TVis['Sex']=='female']
men = TVis[TVis['Sex']=='male']
ax = sns.distplot(women[women['Survived']==1].Age.dropna(), bins=18, label = survived, ax = axes[0], kde =False)
ax = sns.distplot(women[women['Survived']==0].Age.dropna(), bins=40, label = not_survived, ax = axes[0], kde =False)
ax.legend()
ax.set_title('Female')
ax = sns.distplot(men[men['Survived']==1].Age.dropna(), bins=18, label = survived, ax = axes[1], kde = False)
ax = sns.distplot(men[men['Survived']==0].Age.dropna(), bins=40, label = not_survived, ax = axes[1], kde = False)
ax.legend()
_ = ax.set_title('Male')
```
