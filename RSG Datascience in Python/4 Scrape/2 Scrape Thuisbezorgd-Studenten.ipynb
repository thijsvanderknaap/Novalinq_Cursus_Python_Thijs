  handig om snel en efficient data op te slaan (Big-Data).
  
  Zoals wij dictionairies (eigenlijk alle vectoren) gebruiken in Python...

  Ook handig voor het opslaan (en vorm geven) van bestanden.  
  Allerhande bestandsformaten maken op de achtergrond gebruik van een key-value paring systeem voor data opslag.

```python
naam, rating, rammount, btijd, bkosten, mkosten, link = [],[],[],[],[],[],[]


for i in restaurants.find_all('meta', {'itemprop':'ratingValue'}):
    rating.append(i.get('content'))
for i in restaurants.find_all('span', {'class':'rating-total'}):
    rammount.append(i.text)
for i in restaurants.find_all('a', {'class':'restaurantname'}):
    naam.append(i.text)
for i in restaurants.find_all('div', {'class':"avgdeliverytime avgdeliverytimefull open"}):
    btijd.append(i.text)
for i in restaurants.find_all('div', {'class': 'delivery-cost js-delivery-cost'}):
    bkosten.append(i.text)
for i in restaurants.find_all('div', {'class': 'min-order'}):
    mkosten.append(i.text)
for i in restaurants.find_all('a', {'class':'restaurantname'}):
    link.append(i.attrs['href'])
#     restItem = BeautifulSoup(item, 'lxml')
#     print(restItem.findAll('div', {"class":"restaurantname"}))
```

```python
TB = pd.DataFrame([naam, rating, rammount, btijd, bkosten, mkosten, link],index=['naam', 'rating','rating_ammount', 'btijd', 'bkosten', 'mkosten', 'link']).T
```

  Gebruik de functional programming typische **map()** functie.  
  Zoals we hebben laten zien bij de 'title' atribuut van de Titanic.  
  
  Vervang hiermee alle onhandige string-elementen uit attributen die we beter als numeric kunnen gebruiken.  
  
  Vervang dan alle *none* datapunten met een **np.nan**   
  np.nan is beter handelbaar voor veel packetten.  
  Gebruik hiervoor wederom een for loop.

```python

TB.loc[TB['mkosten'].notnull(), 'mkosten'] = list(map(lambda x: x.strip('Min. € '), TB.loc[TB['mkosten'].notnull(), 'mkosten']))
TB.loc[TB['mkosten'].notnull(), 'mkosten'] = list(map(lambda x: x.replace(',', "."), TB.loc[TB['mkosten'].notnull(), 'mkosten']))

TB.loc[TB['bkosten'].notnull(), 'bkosten'] = list(map(lambda x: x.strip(' € '), TB.loc[TB['bkosten'].notnull(), 'bkosten']))

TB.loc[TB['bkosten'].notnull(), 'bkosten'] = list(map(lambda x: x.replace('GRATIS', "0"), TB.loc[TB['bkosten'].notnull(), 'bkosten']))
TB.loc[TB['bkosten'].notnull(), 'bkosten'] = list(map(lambda x: x.replace(',', "."), TB.loc[TB['bkosten'].notnull(), 'bkosten']))

TB.loc[TB['btijd'].notnull(), 'btijd'] = list(map(lambda x : x.strip('min'), TB.loc[TB['btijd'].notnull(), 'btijd']))
TB.loc[TB['btijd'].notnull(), 'btijd'] = list(map(lambda x:x.strip('ca.'), TB.loc[TB['btijd'].notnull(), 'btijd']))

TB.loc[TB['btijd'].notnull(), 'btijd'] = list(map(lambda x:x.strip('Vanaf '), TB.loc[TB['btijd'].notnull(), 'btijd']))

TB.loc[TB['btijd'].notnull(), 'btijd'] = list(map(lambda x:x.replace(':', ","), TB.loc[TB['btijd'].notnull(), 'btijd']))


TB.loc[TB['rating_ammount'].notnull(), 'rating_ammount'] = list(map(lambda x:x.strip('('), TB.loc[TB['rating_ammount'].notnull(), 'rating_ammount']))
TB.loc[TB['rating_ammount'].notnull(), 'rating_ammount'] = list(map(lambda x:x.strip(')'), TB.loc[TB['rating_ammount'].notnull(), 'rating_ammount']))
```
